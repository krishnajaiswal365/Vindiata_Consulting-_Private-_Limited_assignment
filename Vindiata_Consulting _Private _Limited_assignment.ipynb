```python
total_deposits = deposits.groupby('User Id')['Amount'].sum().reset_index()
#total_withdrawals = withdrawals.groupby('User Id')['Amount'].sum().reset_index()
#games = games_played.groupby('User Id')['Games Played'].sum().reset_index()
```

```python
total_withdrawals = withdrawals.groupby('User Id')['Amount'].sum().reset_index()
```

```python
games = games_played.groupby('User ID')['Games Played'].sum().reset_index()
```

```python
user_data = pd.merge(total_deposits,total_withdrawals, on='user id', how='outer')
#user_data = pd.merge(user_data, games, on='user_id', how='outer')
```

```python
user_data.rename({"user id":'User ID'},axis=1,inplace=True)
```

```python
user_data = pd.merge(user_data, games, on='User ID', how='outer')
```

```python
user_data.fillna(0, inplace=True)
```

```python
user_data['Games Played'] = user_data['Games Played'].astype('float64')
```

```python
# Calculate the loyalty points
user_data['loyalty_points'] = (
    0.01 * user_data['deposit amount'] +           # Deposit amount
    0.005 * user_data['withdrawal amount'] +          # Withdrawal amount
    0.001 * user_data[['deposit amount', 'withdrawal amount']].apply(lambda row: max(row[0] - row[1], 0), axis=1) +  # Deposit - Withdrawal
    0.2 * user_data['Games Played']             # Number of games played
)
```

```python
# Rename columns for clarity
user_data.columns = ['user_id', 'total_deposits', 'total_withdrawals', 'num_games', 'loyalty_points']
```

```python
# Step 2: Ranking players by loyalty points
user_data['rank'] = user_data['loyalty_points'].rank(ascending=False, method='min')
```

```python
total_bonus_pool = 10000
top_50_players = user_data[user_data['rank'] <= 50].sort_values(by='rank')
```

```python
top_50_players['bonus'] = total_bonus_pool / 50
```

```python
# Display the result
print(top_50_players[['user_id', 'loyalty_points', 'rank', 'bonus']])
```

```python
from datetime import datetime

def calculate_loyalty_points(entry_time):
    # Parsing entry time to determine the slot
    hour = entry_time.hour

    if 0 <= hour < 12:
        slot = 'S1'
        # Apply points logic for Slot S1 (e.g., 10 points per slot)
        points = 10
    else:
        slot = 'S2'
        # Apply points logic for Slot S2 (e.g., 15 points per slot)
        points = 15

    return slot, points

# Example usage:
entry_time = datetime.now()  # Current time as entry time
slot, points = calculate_loyalty_points(entry_time)

print(f"User checked in during {slot}, earning {points} loyalty points.")
```
